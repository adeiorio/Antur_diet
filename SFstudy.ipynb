```python
import ROOT
from array import array
```

```python
def make_graph_err(entries, x_array, err_x_array, y_array, err_y_array, name, title, x_title, y_title, color, style = 20):
    graph = ROOT.TGraphErrors(entries, x_array, y_array, err_x_array, err_y_array)
    graph.SetNameTitle(name, title)
    graph.GetXaxis().SetTitle(x_title)
    graph.GetYaxis().SetTitle(y_title)
    graph.SetMarkerColor(color)
    graph.SetMarkerStyle(style)
    return graph
```

```python
def SFstudy2vars(tree, clust, branchname1, branchname2, xmin, xmax, gender, optype):
    j=0
    pre_id = -9999
    for entry in tree:
        if entry.Gender != gender:
            continue
        if entry.Local_ID == 42 or entry.Local_ID == 1946:
            continue
        id = entry.Local_ID
        if pre_id != -9999 and pre_id != id and len(x) > 1:
            f1 = ROOT.TF1("f1","pol1", 0, 500)
            if(cluster == clust):
                graph_bmi = make_graph_err(len(x), x, ex, y, ey, "g", "", "Giorni dalla prima visita", branchname1, ROOT.kBlue)
                graph_bmi.Fit("f1")
                if(f1.GetParameter(0) != 0):
                    c2 = ROOT.TCanvas()
                    c2.cd()
                    graph_bmi.Draw()
                    c2.Print("SFplot/SF_id" + str(pre_id) + "_cluster" + str(clust) + "_" + branchname1 + "_" + branchname2 + "_gender" + str(gender) + ".png")
                j += 1
        if pre_id != id:
            cluster = entry.cluster
            pre_id = id
            x = array('f', [])
            ex = array('f', [])
            y = array('f', [])
            ey = array('f', [])
        if(pre_id == id):
            x.append(entry.DeltaDays)
            ex.append(0.5)
            if(cluster == clust):
                if optype == "ratio" or optype == "r":
                    y.append(getattr(entry, branchname1)/getattr(entry, branchname2))
                    ey.append(getattr(entry, branchname1)/getattr(entry, branchname2)*0.03)
                elif optype == 'sub':
                    y.append(getattr(entry, branchname1)-getattr(entry, branchname2))
                    ey.append((getattr(entry, branchname1)-getattr(entry, branchname2))*0.03)

    print(j)
```
