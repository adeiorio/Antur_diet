```python
import ROOT
from array import array
```

```python
ROOT.gStyle.SetOptFit(1)
```

```python
def make_graph_err(entries, x_array, err_x_array, y_array, err_y_array, name, title, x_title, y_title, color, style = 20):
    graph = ROOT.TGraphErrors(entries, x_array, y_array, err_x_array, err_y_array)
    graph.SetNameTitle(name, title)
    graph.GetXaxis().SetTitle(x_title)
    graph.GetYaxis().SetTitle(y_title)
    graph.SetMarkerColor(color)
    graph.SetMarkerStyle(style)
    return graph
```

```python
def SFstudy(tree, id, branchname, gender, percent=False):
    fout = ROOT.TFile("SPplot/" + str(id) + ".root", "UPDATE")
    x = array('f', [])
    ex = array('f', [])
    y = array('f', [])
    ey = array('f', [])
    for entry in tree:
        if entry.Local_ID != id:
            continue
        x.append(entry.DeltaDays)
        ex.append(0.5)
        if(percent):
            y.append(getattr(entry, branchname)*100)
            ey.append(getattr(entry, branchname))
        else:
            y.append(getattr(entry, branchname))
            ey.append(getattr(entry, branchname)*0.01)
        clust = entry.cluster  

    graph_bmi = make_graph_err(len(x), x, ex, y, ey, branchname, "", "Giorni dalla prima visita", branchname, ROOT.kBlue)
    print(len(x))
    f1 = ROOT.TF1("f1","pol1", 0, 500)
    graph_bmi.Fit("f1")
    c2 = ROOT.TCanvas()
    c2.cd()
    graph_bmi.Draw()
    graph_bmi.Write()
    c2.Print("SPplot/SF_id" + str(id) + "_cluster" + str(clust) + "_" + branchname + "_gender" + str(gender) + ".png")
    fout.Close()
```

```python
def SFstudy2vars(tree, id, branchname1, branchname2, gender, optype):
    fout = ROOT.TFile("SPplot/" + str(id) + ".root", "UPDATE")
    x = array('f', [])
    ex = array('f', [])
    y = array('f', [])
    ey = array('f', [])
    for entry in tree:
        if entry.Local_ID != id:
            continue
        x.append(entry.DeltaDays)
        ex.append(0.5)
        if optype == "ratio" or optype == "r":
            if(branchname1 == "BFM" and branchname2 == "SMI"):
                y.append(getattr(entry, branchname1)/getattr(entry, "Weight")/getattr(entry, branchname2)*100)
                ey.append(getattr(entry, branchname1)/getattr(entry, "Weight")/getattr(entry, branchname2))
            else:
                y.append(getattr(entry, branchname1)/getattr(entry, branchname2)*100)
                ey.append(getattr(entry, branchname1)/getattr(entry, branchname2))
        elif optype == 'sub':
            y.append(getattr(entry, branchname1)-getattr(entry, branchname2))
            ey.append((getattr(entry, branchname1)-getattr(entry, branchname2))*0.01)
        clust = entry.cluster  
    f1 = ROOT.TF1("f1","pol1", 0, 500)
    graph_bmi = make_graph_err(len(x), x, ex, y, ey, branchname1 + "_" + branchname2, "", "Giorni dalla prima visita", branchname1, ROOT.kBlue)
    graph_bmi.Fit("f1")
    c2 = ROOT.TCanvas()
    c2.cd()
    graph_bmi.Draw()
    graph_bmi.Write()
    c2.Print("SPplot/SF_id" + str(id) + "_cluster" + str(clust) + "_" + branchname1 + "_" + branchname2 + "_gender" + str(gender) + ".png")
    fout.Close()
```

```python
infile = ROOT.TFile.Open("InBodyExcelData_2022-09-08_18-03-19_ridotto_ML_new.root")
tree = infile.Get("t1")

uomo = 0 #uomini
id_uomo = 1556
```
